```python

def build_prompt(desc):

    # desc = description

    temp =  f"""
    You are a highly experienced software engineer and UML expert with over 15 years in software modeling. Your expertise includes translating natural language requirements into precise UML class diagrams, ensuring syntactic correctness (e.g., proper use of classes, associations, cardinalities), semantic accuracy (e.g., complete representation of domain entities, attributes, methods, and relationships), and pragmatic understandability (e.g., avoiding redundancy and inconsistencies). 
Create a UML class diagram based on the following natural language description. Output the diagram as a detailed textual description, including classes, attributes, methods, and relationships. 
Description: {desc}

"""
    return temp



```

```python

# Gemini API Request
def gemini_caller(prompt):
    headers = {"Content-Type": "application/json"}
    payload = {
        "contents": [{"parts": [{"text": prompt}]}]
    }
    response = requests.post(
        f"{GEMINI_API_URL}?key={GEMINI_API_KEY}",
        headers=headers,
        data=json.dumps(payload)
    )
    if response.status_code == 200:
        candidates = response.json().get("candidates", [])
        if candidates:
            return candidates[0]["content"]["parts"][0]["text"]
    else:
        print("Error:", response.text)
    return None
```

```python

# Interactive Generation Function
def generate_class_diagram(problem_spec):
    prompt = build_prompt(problem_spec)
    diagram = gemini_caller(prompt)
    # evaluation = evaluate_solid(diagram)
    # print("Generated PlantUML Diagram:", diagram)
    
    return diagram
```

```python
import os
import re
from plantuml import PlantUML

server = PlantUML(url="http://www.plantuml.com/plantuml/png/")

# Create folder for outputs
OUTPUT_DIR = "zero-shot-diagrams"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# File to store all UML code
PLANTUML_LOG = os.path.join(OUTPUT_DIR, "zero-shot-plantuml.txt")

def clean_plantuml(text):
    match = re.search(r'@startuml.*?@enduml', text, re.DOTALL)
    return match.group(0).strip() if match else ""

def generate_and_render_diagram(desc, title):
    # Step 1: generate textual UML description
    textual_uml = generate_class_diagram(desc)

    # Step 2: convert textual UML to PlantUML
    new_prompt = f"""
    You will be given a textual description of a system with classes, their attributes, and relationships. 
    The format of the input may vary. Your task is to interpret the description and output a valid PlantUML class diagram. 
    Rules: 
    - Output must start with @startuml and end with @enduml, with nothing else outside.
    - Create a class for each entity with the listed attributes inside it. 
    - Represent relationships (aggregation, composition, associations) using PlantUML notation.
    - Include multiplicities (1, 4, 1..*, *) if mentioned. 
    - If the description mentions that an entity is part of another, model it as aggregation or composition. 
    - Do not add any extra explanation or text — only the PlantUML code.
    {textual_uml}
    """
    plantuml_code = gemini_caller(new_prompt)

    # Step 3: clean PlantUML
    uml_code = clean_plantuml(plantuml_code)

    # print(uml_code)

    # Step 4: render diagram image into the folder
    safe_title = title.replace(" ", "_")
    output_file = os.path.join(OUTPUT_DIR, f"{safe_title}.png")

    try:
        with open(output_file, "wb") as f:
            f.write(server.processes(uml_code))
    except Exception as e:
        print(f"[ERROR] Failed to generate diagram for {title}: {e}")
        return uml_code, None

    # Step 5: append UML code to a log file
    with open(PLANTUML_LOG, "a", encoding="utf-8") as log_file:
        log_file.write("="*60 + "\n")
        log_file.write(f"TITLE: {title}\n")
        log_file.write(f"DESCRIPTION:\n{desc}\n\n")
        log_file.write("PLANTUML CODE:\n")
        log_file.write(uml_code + "\n\n")

    print(f"[OK] Diagram generated for {title}: {output_file}")
    return uml_code, output_file


# Run for first 4 problem descriptions
for item in all_descriptions[4:8]:
    title, desc = item["title"], item["desc"]
    uml_code, diagram_file = generate_and_render_diagram(desc, title)
```
